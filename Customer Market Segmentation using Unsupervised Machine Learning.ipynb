### K-means algorithm

### elbow method

```python
scaler = StandardScaler()
```

```python
data_s = scaler.fit_transform(data)
data.shape
```

```python
data_s
```

```python
scores = []
range_values = range(1,20)

for i in range_values:
    kmean = KMeans(n_clusters = i)
    kmean.fit(data_s)
    scores.append(kmean.inertia_)
    
plt.plot(scores, 'bx-')    
```

```python
kmeans = KMeans(7)
kmeans.fit(data_s)

label = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_,columns=[data.columns])
cluster_centers
```

```python
    cluster_centers = scaler.inverse_transform(cluster_centers)
    
    cluster_centers = pd.DataFrame(data= cluster_centers, columns = [data.columns])
    
    cluster_centers
```

```python
y = kmean.fit_predict(data_s)
y
```

```python
data_cluster = pd.concat([data,pd.DataFrame({'cluster':label})],axis = 1)
```

```python
data_cluster
```

```python
for i in data.columns:
    plt.figure(figsize = (35,5))
    for j in range(7):
        plt.subplot(1
                    ,7,j+1)
        cluster = data_cluster[data_cluster['cluster']==j]
        cluster[i].hist(bins = 20)
    plt.show     
        
```

### PCA

```python
pca = PCA(n_components=2)
principal_comp = pca.fit_transform(data_s)
principal_comp
```

```python
pca_df = pd.DataFrame(data=principal_comp,columns= ['PCA1','PCA2'])
pca_df
```

```python
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':label})],axis = 1)
pca_df
```

```python
plt.figure(figsize = (10,10))
ax = sns.scatterplot(x="PCA1",y="PCA2",hue = "cluster",data = pca_df,palette = ['red','green','blue','yellow','pink','grey','purple'])
plt.show();
```
